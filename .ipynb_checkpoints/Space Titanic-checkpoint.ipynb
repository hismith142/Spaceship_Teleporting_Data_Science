# Would *YOU* Survive the Space Titanic? Let's Find Out!
Using machine learning to determine survival rates of interplanetary travel

### Prompt
"In this competition your task is to predict whether a passenger was transported to an alternate dimension during the Spaceship Titanic's collision with the spacetime anomaly. To help you make these predictions, you're given a set of personal records recovered from the ship's damaged computer system."

### Datafield Descriptions
- PassengerId - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.

- HomePlanet - The planet the passenger departed from, typically their planet of permanent residence.

- CryoSleep - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage. Passengers in cryosleep are confined to their cabins.

- Cabin - The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.

- Destination - The planet the passenger will be debarking to.

- Age - The age of the passenger.

- VIP - Whether the passenger has paid for special VIP service during the voyage.

- RoomService, FoodCourt, ShoppingMall, Spa, VRDeck - Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.

- Name - The first and last names of the passenger.

- Transported - Whether the passenger was transported to another dimension. This is the target, the column you are trying to predict.

## 1. Data Cleaning
Getting the 'train.csv' file to a point where it can be used for a model

```python
import pandas as pd

df = pd.read_csv('space_train.csv')
df.info()
df.describe()
```

Looks like there's some NaN values, let's see how many:

```python
nan_row_count = df.isna().sum() / len(df)
nan_row_count_columns = df.isna().sum().sum() / len(df)
print(nan_row_count)
nan_row_count_columns
```

Each column is missing roughly 2% of its entries, with a total of roughly 25% of rows having some Nan values. Let's see how much we retain with imputation to keep the largest amount of data as possible.

```python
df.head()
```

First, let's handle the missing rows from the HomePlanet column:

```python
df.HomePlanet.unique()
```

```python
df.CryoSleep.unique()
```

```python
def homeplanet(x):
    #ordinally encoding homeplanet data
    homeplanet_map = {'Europa' : 0, 'Earth' : 1, 'Mars' : 2}
    x['HomePlanet'] = x['HomePlanet'].replace(homeplanet_map)
    #imputing values
    x['HomePlanet'].fillna(method='ffill')
    return x
df = homeplanet(df)
df.head()
```

Next, the CryoSleep column:

```python
def cryosleep(x):
    cryosleep_map = {False : 0, True : 1}
    x['CryoSleep'] = x['CryoSleep'].replace(cryosleep_map)
    x['CryoSleep'].fillna(method='ffill')
    return x
df = cryosleep(df)
df.head()
```

Next, the Cabin column. For now, we can just drop it.

```python
def cabin(x):
    x.drop(['Cabin'], axis=1, inplace=True)
    return x
df = cabin(df)
df.head()
```

Then the destination column.

```python
df.Destination.unique()
```

```python
def destination(x):
    destination_map = {'TRAPPIST-1e' : 0, 'PSO J318.5-22' : 1, '55 Cancri e' : 2,}
    x['Destination'].fillna(method='ffill')
    x['Destination'] = x['Destination'].replace(destination_map)
    return x
df = destination(df)
df.head()
```

The VIP column:

```python
def VIP(x):
    x['VIP'] = x['VIP'].fillna(method='ffill')
    vip_map = {True : 1, False : 0}
    x['VIP'] = x['VIP'].replace(vip_map)
    return x
df = VIP(df)
df.head()
```

```python
def purchases(x):
    x['RoomService'].fillna(method='ffill')
    x['FoodCourt'].fillna(method='ffill')
    x['ShoppingMall'].fillna(method='ffill')
    x['Spa'].fillna(method='ffill')
    x['VRDeck'].fillna(method='ffill')
    return x
df = purchases(df)
df.head()
```

```python
def name_and_id(x):
    x.drop(['Name'], axis=1, inplace=True)
    x.drop(['PassengerId'], axis=1, inplace=True)
    return x
df = name_and_id(df)
df.head()
    
```

## Data Correlation Heatmap
What features are related to 'Transported'?

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heatmap = sns.heatmap(df.corr(), cmap='YlGnBu')
heatmap
```

Transported seems to be heavily related to Cryosleep, RoomService, Spa, and VRDeck.

## Creating Model

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
```

```python
#declaring datasets
x = df.drop(['Transported'], axis=1)
y = df.Transported

#building model and scoring over cross-validated data
model = XGBClassifier(n_estimators=100, random_state=1, learning_rate=0.05)
scores = cross_val_score(model, x, y,
                              cv=5,
                              scoring='accuracy')
f"Model is {round(scores.mean(), 4) * 100}% accurate."
```

## Finalizing Model

The model performed with roughly 80% accuracy against our training data cross validation, which is a good start!

```python
test_df = pd.read_csv('space_test.csv')
test_ids = test_df['PassengerId']
```

```python
test_df = homeplanet(test_df)
test_df = cryosleep(test_df)
test_df = cabin(test_df)
test_df = destination(test_df)
test_df = VIP(test_df)
test_df = purchases(test_df)
test_df = name_and_id(test_df)
test_df.head()
```

```python
model.fit(x, y)
final_predicts = model.predict(test_df)
```

```python
submission = pd.DataFrame({
    'PassengerId' : test_ids,
    'Transported' : final_predicts
})
submission.columns
```

```python
submission.head()
```

```python
submission['Transported'].replace({
    0 : False,
    1 : True
}, inplace=True)
```
